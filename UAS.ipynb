```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
url = '/content/drive/MyDrive/Dataset_ImageProcessing/UAS/'
```

```python
img = cv2.imread(url + 'Photo.jpg', 0)
resized_img = cv2.resize(img, (15, 15))

print(resized_img)
```

```python
plt.figure(figsize=(9, 7))

plt.subplot(2, 2, 1);plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB));plt.title('Gambar Original')
plt.subplot(2, 2, 2);plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB));plt.title('Gambar 15 x 15')

plt.show()
```

```python
kernel = np.array(
    [
        [0, 1, 3],
        [1, 2, 1],
        [4, 1, 0]
    ],
    np.uint8
)/10

kernel
```

```python
filtered_resized_img = cv2.filter2D(
    resized_img,
    -1,
    kernel,
    borderType=cv2.BORDER_CONSTANT
)


filtered_resized_img = cv2.copyMakeBorder(
    filtered_resized_img,
    1, 1, 1, 1,
    cv2.BORDER_CONSTANT,
    None,
    value = 0
)


plt.figure(figsize=(9, 7))

plt.subplot(2, 2, 1);plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB));plt.title('Gambar 15 x 15')
plt.subplot(2, 2, 2);plt.imshow(cv2.cvtColor(filtered_resized_img, cv2.COLOR_BGR2RGB));plt.title('Gambar 15 x 15 setelah konvolusi')

plt.show()
```
